```python
#importing required libraries

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver import ActionChains
import time
import pandas as pd


def scroll():      #defining scroll to scroll the page
    scrolling = True 
    while scrolling:
        old_page_source=driver.page_source   # driver.page source used to find the current page source information
        body=driver.find_element(By.TAG_NAME,"body") # using find element by tag name body
        body.send_keys(Keys.END)
        time.sleep(5)
        new_page_source=driver.page_source
        if new_page_source==old_page_source:
            scrolling=False
           


driver = webdriver.Chrome()  # create an instance of the Chrome WebDriver 

Transport = ["https://www.redbus.in/online-booking/rsrtc"] # can add route links

route= []

for i in Transport:
    driver.get(i) # to get specific URL
    driver.maximize_window()
    time.sleep(2)

    scroll()

    pagination_container = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.DC_117_paginationTable"))) # identifying pagetab
    page_tabs = pagination_container.find_elements(By.CSS_SELECTOR, "div.DC_117_pageTabs")
    
    actions= ActionChains(driver)

    for page_tab in page_tabs:  # Acessing page in each pagetab
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable(page_tab))
        actions.move_to_element(page_tab).click().perform()
        time.sleep(5)
        scroll()
        routes = WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div[class='route_details']"))) # collecting route and route link
        for i in routes:
           route_name = i.find_element(By.CSS_SELECTOR, "a[class='route']") 
           route_link = route_name.get_attribute('href')
           route.append((route_name.text, route_link))

all_bus_data=[]

for route_name,route_link in route: 
    try :

        driver.get(route_link)
        time.sleep(2)

        clickables = driver.find_elements(By.CSS_SELECTOR, "div[class='button']") 
        for i in clickables:
            WebDriverWait(driver, 15).until(EC.element_to_be_clickable(i)).click() # Cliciking View buses

        time.sleep(3)
        scroll()
        # using CSS selector to identify elements
        busnames = driver.find_elements(By.CSS_SELECTOR, "div[class='travels lh-24 f-bold d-color']")
        busname_list = [i.text for i in busnames]

        bustypes = driver.find_elements(By.CSS_SELECTOR,"div[class='bus-type f-12 m-top-16 l-color evBus']")
        bustype_list =  [i.text for i in bustypes]
        
        departing_time =driver.find_elements(By.CSS_SELECTOR,"div[class='dp-time f-19 d-color f-bold']")
        departing_time_list=[i.text for i in departing_time]
        
        duration = driver.find_elements(By.CSS_SELECTOR,"div[class='dur l-color lh-24']")
        duration_list = [i.text for i in duration]
        
        reaching_time = driver.find_elements(By.CSS_SELECTOR,"div[class='bp-time f-19 d-color disp-Inline']")
        reaching_time_list = [i.text for i in reaching_time]
        
        star_rating= driver.find_elements(By.CSS_SELECTOR,"div[class='column-six p-right-10 w-10 fl']")
        star_rating_list = [i.text[:3] for i in star_rating]
        
        price =  driver.find_elements(By.CSS_SELECTOR,"span.f-19.f-bold") 
        price_list = [i.text for i in price]
        
        seats_available = driver.find_elements(By.CSS_SELECTOR,"div[class='column-eight w-15 fl']")
        seats_available_list = [i.text[:2] for i in seats_available]

        for busname, bustype, dep_time, dur, reach_time, rating, price, seats in zip(
                busname_list, bustype_list, departing_time_list, duration_list,
                reaching_time_list, star_rating_list, price_list, seats_available_list):
            
            all_bus_data.append({
                'route_name': route_name,
                'route_link': route_link,
                'bus_name': busname,
                'bustype': bustype,
                'departing_time': dep_time,
                'duration': dur,
                'reaching_time': reach_time,
                'star_rating': rating,
                'price': price,
                'seats_available': seats
            })        

        driver.back()      
        time.sleep(5)
        driver.back()   

    except StaleElementReferenceException:
        driver.get(route_link)
        time.sleep(2)

        clickables = driver.find_elements(By.CSS_SELECTOR, "div[class='button']")
        for i in clickables:
            WebDriverWait(driver, 15).until(EC.element_to_be_clickable(i)).click() 

        time.sleep(3)
        scroll()
        
        busnames = driver.find_elements(By.CSS_SELECTOR, "div[class='travels lh-24 f-bold d-color']")
        busname_list = [i.text for i in busnames]

        bustypes = driver.find_elements(By.CSS_SELECTOR,"div[class='bus-type f-12 m-top-16 l-color evBus']")
        bustype_list =  [i.text for i in bustypes]
        
        departing_time =driver.find_elements(By.CSS_SELECTOR,"div[class='dp-time f-19 d-color f-bold']")
        departing_time_list=[i.text for i in departing_time]
        
        duration = driver.find_elements(By.CSS_SELECTOR,"div[class='dur l-color lh-24']")
        duration_list = [i.text for i in duration]
        
        reaching_time = driver.find_elements(By.CSS_SELECTOR,"div[class='bp-time f-19 d-color disp-Inline']")
        reaching_time_list = [i.text for i in reaching_time]
        
        star_rating= driver.find_elements(By.CSS_SELECTOR,"div[class='column-six p-right-10 w-10 fl']")
        star_rating_list = [i.text[:3] for i in star_rating]
        
        price =  driver.find_elements(By.CSS_SELECTOR,"span.f-19.f-bold")
        price_list = [i.text for i in price]
        
        seats_available = driver.find_elements(By.CSS_SELECTOR,"div[class='column-eight w-15 fl']")
        seats_available_list = [i.text[:2] for i in seats_available]


        #using zip file to push elements into list
        for busname, bustype, dep_time, dur, reach_time, rating, price, seats in zip(
                busname_list, bustype_list, departing_time_list, duration_list,
                reaching_time_list, star_rating_list, price_list, seats_available_list):
            
            all_bus_data.append({
                'route_name': route_name,
                'route_link': route_link,
                'bus_name': busname,
                'bustype': bustype,
                'departing_time': dep_time,
                'duration': dur,
                'reaching_time': reach_time,
                'star_rating': rating,
                'price': price,
                'seats_available': seats
            })        

        driver.back()      
        time.sleep(5)
        driver.back() 


driver.quit()
     
           
 
```

```python
import pandas as pd # Converting list to pandas dataframe
df = pd.DataFrame(all_bus_data)
```

```python
#conversion
df['star_rating'] = pd.to_numeric(df['star_rating'],errors='coerce') # converting Str to Numeric in which errors='coerce' move non-nummeric type to NaN
df['price'] = pd.to_numeric(df['price'])
df['seats_available'] = pd.to_numeric(df['seats_available'])
```

```python
import mysql.connector # python module to set-up mysql connection

# connectiing to the server
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="123456789"
)

mycursor = mydb.cursor() # to create a cursor object


mycursor.execute("CREATE DATABASE Redbus") # to execute query
mycursor.execute("USE Redbus")

mycursor.execute("""
    CREATE TABLE busdetails (
        id INT AUTO_INCREMENT PRIMARY KEY,
        Route_name VARCHAR(50),
        Route_link VARCHAR(500),               
        busname VARCHAR(50),
        bustype VARCHAR(50),
        departing_time TIME,
        duration VARCHAR(50),
        reaching_time TIME,
        star_rating FLoat,
        price DECIMAL(10, 2),
        seats_available INT
    )
""")

 #pushing data into sql

sql = """
    INSERT INTO busdetails (Route_name, Route_link, busname, bustype, departing_time, duration, reaching_time, star_rating, price, seats_available)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for index, row in df.iterrows():
    val = (
        row['route_name'], row['route_link'], row['bus_name'], row['bustype'], row['departing_time'],
        row['duration'], row['reaching_time'], row['star_rating'], row['price'], row['seats_available']
    )
    mycursor.execute(sql, val)

mydb.commit()


mycursor.close()
mydb.close()
```
